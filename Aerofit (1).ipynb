**2.For each AeroFit treadmill product, construct two-way contingency tables and compute all conditional and marginal probabilities along with their insights/impact on the business.**

```python
contingency_table = pd.crosstab(dsp['Product'], dsp['Gender'])
print(contingency_table)
```

```python
conditional_crosstab = pd.crosstab(dsp['Product'], dsp['Gender'], margins=True, normalize=True)
print("Conditional Crosstab (Probabilities based on Gender):\n", conditional_crosstab)
```

*  **Insights**

   **Product KP281 is equally popular among both genders, while KP781 is predominantly used by males.**

  **Overall,males has a larger portion of product users (57.78%) compared to females (42.22%).**

```python
marginal_crosstab = pd.crosstab(dsp['Product'], dsp['Gender'],margins=True,normalize=True)
print("\nMarginal Crosstab (Overall Probabilities):\n", marginal_crosstab)
```

* **Insights**

   **KP481 Threadmill shows a balanced usage between genders,contributing to 33.33% of overall usage.**
